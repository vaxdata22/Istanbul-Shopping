####   Phase 4: Data Minning (Uncovering Insights)

Now we need to attend to the data analysis questions for more inisgts.

We need to carry out some data manipulation on the following information:
  
* which product categories generates the highest revenue for the shopping mall?

* How does the gender distribution of shoppers varies across different categories?

* what are the most popular products within each categories based on quantity sold?

* what are the average transaction per age group?

* what are the total transaction per age group

* what are the most preferred payment method for each age group?

* what are the most preferred payment method for each gender?

* what are the specific shopping malls that attract more male or female shoppers?




```python
##  Let's add new columns that we will be needing for our Business Analysis

df['total_sales'] = df['Quantity'] * df['Price']

df
```

```python
# we can use agg function to determine the maximum & mimimum sales

df.total_sales.agg(['min','max'])
```

```python
# Now let's check for the statistics of our data one more time since a new column 'total sales' have been added to our dataset.

df.describe()
```

```python
# let's check for the  product category that generates the highest revenue from2021-2023

highest_revenue_cat = df.groupby('Category', as_index=False)[['total_sales']].sum().sort_values('total_sales', ascending=False)

highest_revenue_cat
```

```python
# How does the gender distribution of shoppers varies across different categories? 

Gender_distribution = df.groupby('Category',as_index=False)[['Gender']].value_counts()\
            .sort_values('Category',ascending=False).rename(columns={'count':'total_gender_population_per_category'})

Gender_distribution
```

#####  Gender Distribution Across Categories:

As seen in the visualization for various categories, Female gender is leading in terms of shopping for all categories

```python
# what are the most popular categories based on quantity sold?

cat_quantity = df.groupby('Category',as_index=False)[['Quantity']].sum().sort_values('Quantity',ascending=False)

cat_quantity
```

```python
# what are the average transaction per each age group?

Avg_transaction = df.groupby('Age', as_index=False)[['total_sales']].mean().sort_values('Age', ascending=False)

Avg_transaction
```

```python
 # what are the most preferred payment method for each Age?
    
age_payment = df.groupby(['Age', 'Payment_Method'], as_index=False)[['Payment_Method']].value_counts()\
        .sort_values('Payment_Method',ascending=True).rename(columns={'count':'total_payment_per_age'})

age_payment
```

```python
# Now let's check and see the most commonly used payment method

df['Payment_Method'].value_counts().idxmax()
```

```python
# Now let's check and be sure if cash is the most prefered payment  

prefered_payment_method = df[['Payment_Method']].value_counts()

prefered_payment_method
```

```python
# what are the most preferred payment method for each gender?
    
Gender_preferred_payment_method = df.groupby('Gender',as_index=False)[['Payment_Method']] \
                                        .value_counts().rename(columns={'count':'total_payment_per_gender'})

Gender_preferred_payment_method
```

```python
# what are the specific shopping malls that attract more male or female shoppers?

male_female_shoppers = df.groupby(['Shopping_Mall', 'Gender'], as_index=False)[['Shopping_Mall']].value_counts() \
.rename(columns={'count':'total_pop_for_male_female_shoppers'}).sort_values(['total_pop_for_male_female_shoppers'], ascending=False)

male_female_shoppers
```

```python
# it can be clearly seen that women are shopping more than men

shoppers_population = df.groupby('Gender',as_index=False)[['Shopping_Mall']].count().sort_values(['Shopping_Mall'], ascending=False) \
            .rename(columns={'Shopping_Mall': 'total_no_of_male_female_shoppers'})

shoppers_population
```

#### Phase 5: Data visualization

```python
# let's visualize for the  product category that generates the highest revenue from 2021-2023?

plt.figure(figsize=(20,10))

sns.set_style("whitegrid")

sns.barplot(data=highest_revenue_cat, y='Category', x = 'total_sales')

plt.show()
```

```python
#  let's visualize for the  product category that generates the highest revenue using pie 
plt.pie(x=highest_revenue_cat['total_sales'], labels=highest_revenue_cat['Category'], radius=2.5, \
                autopct = '%.2f%%', labeldistance=1.02)
plt.show()
```

```python
# let's us visualize how the gender distribution of shoppers varies acrosss shopping categories from 2021-2023

plt.figure(figsize=(20,10))

sns.set_style("whitegrid")

sns.barplot(data=Gender_distribution, x='Category', y= 'total_gender_population_per_category', hue ='Gender')

plt.show()
```

```python
# let's visualize for the Most popular category based on quantity sold between 2021-2023

plt.figure(figsize=(20,10))

sns.set_style("whitegrid")

sns.barplot(data=cat_quantity, y='Category', x= 'Quantity')

plt.show()
```

```python
#  let's visualize for the Most popular category based on quantity sold between 2021-2023 using pie 

plt.pie(x=cat_quantity['Quantity'], labels=cat_quantity['Category'], radius=2.5, \
                autopct = '%.2f%%', labeldistance=1.02)

plt.show()
```

```python
# Visualization for each age and their average sales recorded between 2021-2023

plt.figure(figsize=(20,10))

sns.set_style("whitegrid")

sns.barplot(data=Avg_transaction, x='Age', y= 'total_sales')

plt.show()
```

```python
#  let's visualize the Most preferred payment Method per each age group

plt.figure(figsize=(25,10))

sns.barplot(data=age_payment, x = 'Age', y= 'total_payment_per_age', hue='Payment_Method')

plt.show()
```

```python
#  let's visualize the Most preferred payment Method.

prefered_payment_method.plot(kind='pie', radius =2.5, autopct = '%.2f%%', labeldistance=1.02)

plt.show()
```

```python
# let's visualize for The most preferred payment across all Gender

plt.figure(figsize=(20,10))

sns.set_style("whitegrid")

sns.barplot(data=Gender_preferred_payment_method, x='Gender', y= 'total_payment_per_gender', hue ='Payment_Method')

plt.show()
```

```python
# let's visualize for male and female_shoppers across shopping malls

plt.figure(figsize=(16,8))

sns.barplot(data=male_female_shoppers, x = 'Gender', y= 'total_pop_for_male_female_shoppers', hue='Shopping_Mall')

plt.show()
```

```python
# let's visualize for thr total number of Genders and how they shopped.

plt.figure(figsize=(16, 8))

sns.barplot(data=shoppers_population, x='Gender', y ='total_no_of_male_female_shoppers', palette=['red','green'])

plt.show()
```

```python
# To see the visual correlation between 'Quantity' and 'price' in the shopping mall datset from 2021-2023

plt.figure(figsize=(25, 10))

sns.regplot(data=df, x='Quantity', y='Price')

plt.show()
```

```python
# Checking for the correlation coefficient between the numerical fields in the shopping mall datset from 2021-2023

df.corr(method='pearson')[['total_sales']]
```

#####  A Finding From The Correlation Above Shows That There is :

*  A postive correlation between total sales and price and the Value = 0.962372
